```python
from MLPmodules import *
```

```python
config_claasification = MLPConfig(
    task="classification",
    hidden_layers=[64, 32],
    activation="leakyrelu",
    output_activation="softmax",
    optimizer="AdamW",
    learning_rate=0.001,
    batch_size=32,
    epochs=100,
    dropout=0.2,
    l2=0.001,
    test_size=0.2,
    random_state=42,
    scaling="standard",
    loss="categorical_crossentropy",
    metric = "accuracy",
    name="LeakyReLU + AdamW"
)
```

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
           'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
data = pd.read_csv(url, header=None, names=columns)

# Glance of the data
data.head()
```

```python
X = data.drop('Outcome', axis=1).values
y = data['Outcome'].values
unique_class = len(set(y))

exp = Experiment(X, y, configs=[config_claasification], n_classes=unique_class, result_file="result1.csv")
```

```python
exp.run()
```
